#### Import Libraries

```python
from pdf2image import convert_from_path
import cv2
import pytesseract
import imutils
import numpy as np
import gc 
```

```python
#set path to tesseract 
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

#### Convert PDF to images

```python
#converting pdf pages to image (1-100)
files =  convert_from_path(pdf_path= r'.\PDF\Thoreau_All.pdf', 
                          dpi=100,
                          first_page=4000,#8001,
                          last_page=10531,#10531,
                          output_folder= r'.\Images',
                          output_file='Thoreau',
                          fmt ='jpg',
                          transparent=True
                         )
```

```python
#pdf page 33 converted to image
files[33]
```

#### Run OCR on the Images

```python
# in order to apply Tesseract v4 to OCR text we must supply
# (1) a language, 
# (2) an OEM flag of 4, indicating that the we wish to use the LSTM neural net model for OCR, and finally
# (3) an OEM value, in this case, 7 which implies that we are treating the ROI as a single line of text
config = r'--oem 1 --psm 7'
text = pytesseract.image_to_string(files[33],lang='eng')
```

```python
#Converted Text (Page 33)
print(text)
```

```python
#Create a text file
file = open(r'.\Text\Thoreau.txt','a') 

for f in files:
    text = pytesseract.image_to_string(f,lang='eng')
    file.write(text)

file.close() 
```
